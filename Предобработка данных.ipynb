```python
#заменим пустые значения в этих типах дохода медианными
gov_median = df.loc[df['income_type'] == 'госслужащий', 'total_income'].median()
df.loc[df['income_type'] == 'госслужащий', 'total_income'] = df.loc[df['income_type'] == 'госслужащий', 'total_income'].fillna(gov_median)

companion_median = df.loc[df['income_type'] == 'компаньон', 'total_income'].median()
df.loc[df['income_type'] == 'компаньон', 'total_income'] = df.loc[df['income_type'] == 'компаньон', 'total_income'].fillna(companion_median)

pension_median = df.loc[df['income_type'] == 'пенсионер', 'total_income'].median()
df.loc[df['income_type'] == 'пенсионер', 'total_income'] = df.loc[df['income_type'] == 'пенсионер', 'total_income'].fillna(pension_median)

business_median = df.loc[df['income_type'] == 'предприниматель', 'total_income'].median()
df.loc[df['income_type'] == 'предприниматель', 'total_income'] = df.loc[df['income_type'] == 'предприниматель', 'total_income'].fillna(business_median)

employee_median = df.loc[df['income_type'] == 'сотрудник', 'total_income'].median()
df.loc[df['income_type'] == 'сотрудник', 'total_income'] = df.loc[df['income_type'] == 'сотрудник', 'total_income'].fillna(employee_median)

# проверим, остались ли пропуски
df['nan_income'] = df['total_income'].isna()
df.groupby('income_type').agg({'total_income': 'count', 'nan_income': 'sum'}).astype(int)
```

### Избавляемся от артефактов


Я обратил внимание, что в столбце children есть аномальные значения: -1 и 20. Можно предположить, что -1 это на самом деле 1 со случайно напечатанным минусом. Насчет цифры 20 может быть два варианта: 1) это 2 со случайно напечатанным нулём 2) у человека действительно 20 детей: например своих и приёмных. Поменяем -1 на 1 везде. Клиентов со значением 20 в строке children уберём из данных (они составляют всего 0,3% от всех клиентов).

```python
# очистим данные от тех клиентов, которым приписывается наличие 20 детей
df = df[df['children'] != 20]
#Заменим значения -1 на 1
df.loc[df["children"] == -1,"children"] = 1
# проверим, получилось ли
df['children'].value_counts()
```

### Лемматизация

```python
# найдем наиболее часто употребляемые в целях существительные - одновременно цели кредитования
from collections import Counter
from pymystem3 import Mystem
m = Mystem()
lemm_purposes = Counter(m.lemmatize(' '.join(df['purpose'])))
lemm_purposes
```

```python
# лемматизируем значения столбца purpose
df['purpose'] = df['purpose'].apply(m.lemmatize)
# напишем функцию, которая поможет сократить цель кредита до одного слова
def one_purpose(row):
    for purpose in row:
        if 'свадьба' in row:
            return 'свадьба'
        if 'образование' in row:
            return 'образование'
        if 'недвижимость' in row:
            return 'недвижимость'
        if 'жилье' in row:
            return 'недвижимость'
        if 'автомобиль' in row:
            return 'автомобиль'
        return 'прочее'
df['new_purpose'] = df['purpose'].apply(one_purpose)
df['new_purpose'].value_counts()
```

Всего я выделил четыре главных цели: недвижимость, образование, автомобиль и свадьба. Жилая и коммерческая недвижимость объединены. ### Категоризация данных

```python
# создадим словарь, в который поместим столбцы с доходом и наличием задолженности
income_dict = df[['total_income','debt']]
# удалим из словаря дубликаты и сбросим индексы
income_dict = income_dict.drop_duplicates().reset_index(drop=True)
df_income_grouped = income_dict.groupby('total_income').count()
# Напишем функцию для разделения клиентов по размеру доходов
def income_group(total_income):
    if total_income <= 100000:
        return 'Доход до 100 тысяч рублей'
    if total_income <= 200000:
        return 'Доход 100-200 тысяч рублей'
    return 'Доход более 200 тысяч рублей'
# добавим в словарь столбец с названиями групп, на которые разбили клиентов
income_dict['income_groups'] = income_dict['total_income'].apply(income_group)
income_dict.head()
```

```python
# создадим словарь, в который поместим цель кредита, наличие долга и общий доход.
#Столбец с доходами пригодится, чтобы при создании сводной таблицы по этому столбцу считались строки
purpose_dict = df[['new_purpose','debt', 'total_income']]
# удалим из словаря дубликаты и сбросим индексы
purpose_dict = purpose_dict.drop_duplicates().reset_index(drop=True)

purpose_dict.head()
```

```python
# создадим словарь, в который поместим столбцы с количеством детей, наличием долга и доходом 
children_dict = df[['children','debt', 'total_income']]
# удалим из словаря дубликаты и сбросим индексы
children_dict = children_dict.drop_duplicates().reset_index(drop=True)
df_children_grouped = children_dict.groupby('children').count()
# напишем функцию, которая поможет сгруппировать людей на тех, у кого дети есть и у кого их нет
def children_group(children):
    if children > 0:
        return 'Есть дети'
    return 'Детей нет'
children_dict['children_groups'] = children_dict['children'].apply(children_group)
children_dict.head()
```

```python
# создадим словарь, в который поместим столбцы с семейным статусом, доходом и наличием долга
family_dict = df[['family_status','debt', 'total_income']]
# удалим из словаря дубликаты и сбросим индексы
family_dict = family_dict.drop_duplicates().reset_index(drop=True)

family_dict.head()
```

### Вывод

Чтобы в следующем разделе ответить на четыре вопроса, мы создали четыре словаря. Теперь у нас есть отдельные датафреймы для тех, у кого есть дети и укого их нет, для людей с разным уровнем дохода, семейным статусом и целью кредита. В эти словари я также поместил данные о наличии задолженности и месячном доходе. Имея четыре таких словаря, мы сможем ответить на вопрос о зависимости просрочки от вышеперечисленных данных (наличие детей, семейный статус, цель кредита и уровень дохода).

Посколько нам нужно ответить на вопрос "Есть ли зависимость между наличием детей и возвратом кредита в срок?", то точное количество детей неважно. Поэтому я оставил только две группы: есть дети и детей нет. По размерам дохода я резделил клиентов на три более или менее равные группы.

### Шаг 3. Ответьте на вопросы

- Есть ли зависимость между наличием детей и возвратом кредита в срок?

```python
# Сделаем сводную таблицу 
children_pivot = children_dict.pivot_table(index = ['children_groups'], 
                                columns = 'debt', 
                                values = 'total_income', 
                                aggfunc = 'count').fillna(0)
# у тех, кто имеет 5 детей, было NaN там, где должен быть 0,
#поэтому я заменил отсутствующие значения на 0.
new = ['Задолженность отсутствует', 'Была задолженность']
children_pivot.set_axis(new, axis = 'columns', inplace = True)
children_pivot['%'] = children_pivot['Была задолженность'] / children_pivot['Задолженность отсутствует'] * 100

children_pivot
```

### Вывод

Зависимость есть. 8,3% из тех, у кого нет детей, имели просрочку по кредиту. Среди тех, у кого дети есть, просрочки встречаются чаще: таких здесь 10,4%.

- Есть ли зависимость между семейным положением и возвратом кредита в срок?

Среди женатых/замужних клиентов просрочивших на 3 процентных пункта меньше (8,4%). Ещё меньше вероятность просрочки по кредиту у вдовцов (7,3%) и разведённых клиентов (7,7%).
